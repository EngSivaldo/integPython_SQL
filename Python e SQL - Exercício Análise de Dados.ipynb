### Análise de Dados


```python
# filtrar tabela - > garantindo que estamos só com San Francisco
tabela_salarios = tabela_salarios.loc[tabela_salarios["Agency"]== "San Francisco", :]
display(tabela_salarios)
```

```python
import pandas as pd
import numpy as np

# Converter colunas relevantes para numéricas, caso necessário
tabela_salarios['TotalPay'] = pd.to_numeric(tabela_salarios['TotalPay'], errors='coerce')
tabela_salarios['TotalPayBenefits'] = pd.to_numeric(tabela_salarios['TotalPayBenefits'], errors='coerce')

# Selecionar apenas colunas numéricas para o cálculo da média
colunas_numericas = tabela_salarios.select_dtypes(include=[np.number])

# Incluir a coluna 'Year' nas colunas numéricas para fazer o groupby
colunas_numericas['Year'] = tabela_salarios['Year']

# Agrupar por 'Year' e calcular a média
tabela_sm = colunas_numericas.groupby('Year').mean().reset_index()

# Mostrar o resultado
display(tabela_sm)
```

##### 3. Qual foi a evolução do total gasto com salário ao longo dos anos


```python
tabela_salarios['TotalPay'] = pd.to_numeric(tabela_salarios['TotalPay'], errors='coerce')
tabela_salarios['TotalPayBenefits'] = pd.to_numeric(tabela_salarios['TotalPayBenefits'], errors='coerce')
tabela_salarios.fillna(0, inplace=True)

```

```python
import pandas as pd
import numpy as np

# Definir a função de formatação
def formatar(valor):
    return 'R${:,.2f}'.format(valor)

# Converter colunas para numéricas, se necessário
tabela_salarios['TotalPay'] = pd.to_numeric(tabela_salarios['TotalPay'], errors='coerce')
tabela_salarios['TotalPayBenefits'] = pd.to_numeric(tabela_salarios['TotalPayBenefits'], errors='coerce')

# Preencher valores NaN com 0 (ou use outra forma de tratamento se preferir)
tabela_salarios.fillna(0, inplace=True)

# Selecionar apenas colunas numéricas para o cálculo da soma
colunas_numericas = tabela_salarios.select_dtypes(include=[np.number])

# Incluir a coluna 'Year' nas colunas numéricas para fazer o groupby
colunas_numericas['Year'] = tabela_salarios['Year']

# Agrupar por 'Year' e somar os valores de 'TotalPay' e 'TotalPayBenefits'
tabela_total = colunas_numericas.groupby('Year')[['TotalPay', 'TotalPayBenefits']].sum().reset_index()

# Aplicar a formatação aos valores
tabela_total['TotalPay'] = tabela_total['TotalPay'].apply(formatar)
tabela_total['TotalPayBenefits'] = tabela_total['TotalPayBenefits'].apply(formatar)

# Mostrar o resultado formatado
display(tabela_total)
```
